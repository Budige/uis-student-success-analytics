```python
import pandas as pd
import matplotlib.pyplot as plt

# Load enrollment data
df = pd.read_csv('../data/raw/uis_enrollment_2014_2023.csv')
print(df.describe())
```

```python
# Quick look at enrollment decline
# NOTE: COVID years (2020-2021) show accelerated decline
df.plot(x='year', y='total_enrollment', kind='bar', figsize=(10,5))
plt.title('UIS Total Enrollment 2014-2023')
plt.show()
```
