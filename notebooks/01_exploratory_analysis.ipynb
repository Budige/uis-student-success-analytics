```python
import pandas as pd
import matplotlib.pyplot as plt

# Load enrollment data
df = pd.read_csv('../data/raw/uis_enrollment_2014_2023.csv')
print(df.describe())
```

```python
# Quick look at enrollment decline
# NOTE: COVID years (2020-2021) show accelerated decline
df.plot(x='year', y='total_enrollment', kind='bar', figsize=(10,5))
plt.title('UIS Total Enrollment 2014-2023')
plt.show()
```

```python
# Correlation: Pell % vs retention rate across institutions
# Using benchmark data to test if access mission correlates with lower outcomes
import pandas as pd
bench = pd.read_csv("../data/raw/illinois_universities_comparison_2023.csv")
corr = bench[["pell_grant_pct","retention_rate","grad_rate_6yr"]].corr()
print("Correlation matrix:")
print(corr.round(3))
print()
# Interesting finding: UIS grad enrollment share held steady while UG declined
enroll = pd.read_csv("../data/raw/uis_enrollment_2014_2023.csv")
enroll["grad_pct"] = enroll["grad_enrollment"] / enroll["total_enrollment"] * 100
print("Graduate % of total enrollment by year:")
print(enroll[["year","grad_pct"]].to_string(index=False))
print()
print("Net change: UG declined", enroll.iloc[-1]["undergrad_enrollment"] - enroll.iloc[0]["undergrad_enrollment"],
      "but grad declined only", enroll.iloc[-1]["grad_enrollment"] - enroll.iloc[0]["grad_enrollment"])
```
